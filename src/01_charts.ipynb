**Exercise**

Jill designs solar panels as a hobby
- On April 1st, Jill's "Mark I" design begins generating power: 1 kJ/day.
- On May 1st, her "Mark II" design begins generating 4 kJ of
power per day.

1. What day is it when Jill's Mark II design has generated as much total energy as the Mark I design?
2. How much total energy have both generated by that day?
3. What would be the solutions to 1 and 2 be if Mark II design generated 1kJ of power per day?

```python
# Just plot two points beacause these are linear equations
time = np.linspace(0, 50, 2)

# 30 days from April 1st to May 1st
dm1 = 1 * time
dm2 = 4 * (time - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Power generated')
plt.xlabel('day')
plt.ylabel('total power (kJ)')

ax.plot(time, dm1, c='blue', label='Mark I')
ax.plot(time, dm2, c='red', label='Mark II')
ax.set_ylim(0, 75)

plt.axvline(x=40, color='purple', linestyle='--')
plt.axhline(y=40, color='purple', linestyle='--')

ax.legend()
```

```python
# If Mark2 would generated 1kJ of power per day
dm2 = 1 * (time - 30)

fig, ax = plt.subplots()
plt.title('Power generated')
plt.xlabel('day')
plt.ylabel('total power (kJ)')

ax.plot(time, dm1, c='blue', label='Mark I')
ax.plot(time, dm2, c='red', label='Mark II')

plt.axvline(x=40, color='purple', linestyle='--')
plt.axhline(y=40, color='purple', linestyle='--')

ax.legend()
```

**Solution**
1. Day 40.
2. 40 kJ each one.
3. Never because Mark I and Mark II generates the same power per day.
